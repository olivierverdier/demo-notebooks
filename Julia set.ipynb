Eksempel med Julia-mengde                               

```python
mu = array(-0.02 + 0.01*1j)
N = 100
maxIterations = 50
J = zeros((2*N+1, 2*N+1)) # a square matrix of sixe 2N+1
for x in range(-N,N):
    for y in range(-N,N):
        z = x/N + 1j*y/N
        for i in range(maxIterations):
            if abs(z) > 2:
                break
            z = z**2 + mu/z**2
        J[x+N,y+N] = i*255/50
imshow(J)
```
